```python
import pandas as pd

combined_df = pd.read_csv("combined_data.csv")
hotel_df = pd.read_csv("hotel_reviews.csv")
hotel_df
```

```python
hotel_df = hotel_df[["reviews", "labels"]]

hotel_df.columns = [
    "Sentences_clean",
    "unified_sentiment"
]
hotel_df["domain_name"] = "Hotels"
hotel_df
```

```python
hotel_df = hotel_df[
    [
        "domain_name",
        "unified_sentiment",
        "Sentences_clean"
    ]
]
hotel_df.dropna(inplace=True)

combined_df.dropna(
    subset=["Sentences_clean", "unified_sentiment"],
    inplace=True
)
hotel_df.drop_duplicates(
    subset=["Sentences_clean"],
    inplace=True
)

combined_df.drop_duplicates(
    subset=["Sentences_clean"],
    inplace=True
)
```

```python
print("Original Combined Dataset:")
print(combined_df["unified_sentiment"].value_counts())
print("\nHotel Dataset:")
print(hotel_df["unified_sentiment"].value_counts())
```

```python
final_df = pd.concat(
    [combined_df, hotel_df],
    ignore_index=True
)
```

```python
final_df.drop_duplicates(
    subset=["Sentences_clean"],
    inplace=True
)

final_df.reset_index(
    drop=True,
    inplace=True
)
print("\nFinal Dataset Distribution:")
print(final_df["unified_sentiment"].value_counts())
```

```python
final_df.to_csv(
    "combined_with_hotels.csv",
    index=False,
    encoding="utf-8-sig"
)

print("\nMerged dataset saved successfully!")
```
