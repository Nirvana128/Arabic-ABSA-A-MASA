```python
def get_label(row):
  if(row[2] in ['1','2']):
    return -1
  elif(row[2] in ['3']):
    return 0
  else:
    return 1
```

```python
reviews = df.iloc[:,6]
labels = df.apply(get_label, axis=1)
new_Data = pd.DataFrame()
new_Data['reviews'] = reviews
new_Data['labels'] = labels
new_Data
```

```python
negative_reviews = new_Data[new_Data['labels'] == -1].sample(n=5000, random_state=42)
positive_reviews = new_Data[new_Data['labels'] == 1].sample(n=5000, random_state=42)

final_data = pd.concat([negative_reviews, positive_reviews])
final_data = final_data.sample(frac=1, random_state=42).reset_index(drop=True)

display(final_data.head())
```

```python
print(f"Shape of the final data: {final_data.shape}")
print("Distribution of labels in final data:")
display(final_data['labels'].value_counts())
```

```python
detected_aspects_list = []
for index, row in df.iterrows():
    review_text = str(row['reviews'])
    found_aspects = []
    for word in review_text.split():
        for aspect_name, aspect_details in hotel_aspects.items():
            if word in aspect_details['synonyms']:
                found_aspects.append(word)
                break
    detected_aspects_list.append(found_aspects)

df['detected_aspects'] = detected_aspects_list

display(df.head())
```

```python
df = df[df['detected_aspects'].apply(lambda x: len(x) > 0)]
display(df.head())
```

```python
import re
from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis", model="CAMeL-Lab/bert-base-arabic-camelbert-da-sentiment")

def get_numerical_aspect_sentiment(row):
    review_text = str(row['reviews'])
    segments = re.split(r'[.\u0640\u060C!\?\n]|\sلكن\s', review_text)
    detected_words = row['detected_aspects']

    sentiment_scores = []

    for word in detected_words:
        relevant_segment = next((s for s in segments if word in s), review_text)

        if len(relevant_segment.strip()) > 2:
            result = sentiment_pipeline(relevant_segment[:512])[0]
            label = result['label'].lower()

            if 'pos' in label:
                sentiment_scores.append(1)
            elif 'neg' in label:
                sentiment_scores.append(-1)
            else:
                sentiment_scores.append(0)
        else:
            sentiment_scores.append(0)

    return sentiment_scores
sample_df = df.head(10).copy()
sample_df['aspect_sentimentality'] = sample_df.apply(get_numerical_aspect_sentiment, axis=1)
display(sample_df[['reviews', 'detected_aspects', 'aspect_sentimentality']])
```
