```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import RandomOverSampler
from collections import Counter

# Create a temporary DataFrame to ensure 'unified_sentiment' is numeric and without NaNs
# This is a robust way to handle potential inconsistencies if previous cells were not run correctly or values were overwritten.
df = df.copy()

# Ensure 'unified_sentiment' is truly numeric, coercing any non-numeric (like '') to NaN
df['unified_sentiment'] = pd.to_numeric(df['unified_sentiment'], errors='coerce')

# Drop rows where 'unified_sentiment' is NaN, ensuring only valid sentiment values remain
df = df.dropna(subset=['unified_sentiment'])

# Now separate features (X) and target (y) from the cleaned temporary DataFrame
X_text = df['Sentences']
y = df['unified_sentiment'].astype(int) # Convert to int for classification tasks, now safe as NaNs are removed

print("Original class distribution:")
print(y.value_counts())
print("\n")

# Vectorize the text data - this is a necessary step before applying samplers.
# For demonstration purposes, we vectorize here. In a full pipeline,
# this would be part of the preprocessing steps before the sampler.
vectorizer = TfidfVectorizer(max_features=5000) # Limiting features for speed
X_vectorized = vectorizer.fit_transform(X_text) # X_text is now aligned with y

print(f"Shape of vectorized X before resampling: {X_vectorized.shape}")

# Initialize RandomOverSampler
ros = RandomOverSampler(random_state=42)

# Apply oversampling
X_resampled, y_resampled = ros.fit_resample(X_vectorized, y)

print("Resampled class distribution using RandomOverSampler:")
print(pd.Series(y_resampled).value_counts())
print(f"Shape of X after resampling: {X_resampled.shape}")

print("\nNote: Oversampling/undersampling techniques are typically applied during the training phase of a model. For robust evaluation, ensure that these techniques are applied only to the training data and integrated within a scikit-learn pipeline to prevent data leakage.")
```

**Normalization**

```python

def normalize_arabic(text):
    if not isinstance(text, str):
        return text

    # Remove diacritics (Tashkeel / Harakat) U+064B..U+065F, U+0670
    text = re.sub(r'[\u064B-\u065F\u0670]', '', text)

    # Normalise Alef variants to plain Alef U+0627
    text = re.sub(r'[\u0622\u0623\u0625\u0671]', '\u0627', text)

    # Normalise Teh Marbuta U+0629 to Heh U+0647
    text = re.sub(r'\u0629', '\u0647', text)

    # Normalise dotless Yeh / Alef Maqsura U+0649 to Yeh U+064A
    text = re.sub(r'\u0649', '\u064A', text)

    # Normalise Waw with Hamza U+0624 to plain Waw U+0648
    text = re.sub(r'\u0624', '\u0648', text)

    # Remove Tatweel (kashida) U+0640
    text = re.sub(r'\u0640', '', text)

    # Keep Arabic-script code-points, spaces, and digits; strip everything else
    text = re.sub(r'[^\u0600-\u06FF\u0750-\u077F\u08A0-\u08FF\s\d]', ' ', text)

    # Collapse multiple spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text

df['Sentences_clean'] = df['Sentences'].apply(normalize_arabic)
df['Aspects_clean']   = df['Aspects'].apply(lambda x: normalize_arabic(x) if pd.notna(x) else x)

print("Original  :", df['Sentences'].iloc[0])
print("Normalised:", df['Sentences_clean'].iloc[0])
```

**Parse Aspects Sentiment into a Python list**

```python
def parse_aspect_sentiments(raw):
    '''Converts '1, 1, -1' into [1, 1, -1]. Returns None on failure.'''
    if not isinstance(raw, str):
        return None
    try:
        return [int(v.strip()) for v in raw.split(',')]
    except ValueError:
        return None

df['Aspects_sentiments_list'] = df['Aspects Sentiment'].apply(parse_aspect_sentiments)
print("Sample parsed aspect sentiments:")
print(df[['Aspects Sentiment', 'Aspects_sentiments_list']].dropna(subset=['Aspects Sentiment']).head(5))
```

**Parse Aspects into a Python list of aspect terms***

```python
def parse_aspect_list(raw):
    '''Splits an Arabic comma-separated string into a Python list.'''
    if not isinstance(raw, str):
        return None
    # Arabic separator is U+060C; Western comma is ,
    parts = re.split(r'[،,]', raw)
    parts = [p.strip() for p in parts if p.strip()]
    return parts if parts else None

df['Aspects_list'] = df['Aspects_clean'].apply(parse_aspect_list)
print("Sample parsed aspect lists:")
print(df[['Aspects_clean', 'Aspects_list']].dropna(subset=['Aspects_clean']).head(5))
```

### Heatmap of Aspect Sentiments

This heatmap visualizes the count of each aspect across different sentiment categories (-1: Negative, 0: Neutral, 1: Positive). It helps to quickly identify aspects that frequently receive a particular sentiment.

```python
import arabic_reshaper
from bidi.algorithm import get_display

# Function to extract individual aspects and their corresponding sentiments
def extract_aspect_sentiment_pairs(df: pd.DataFrame):
    aspect_sentiment_pairs = []
    for index, row in df.iterrows():
        aspects_str = row['Aspects_clean']
        sentiment_str = row['Aspects_sentiments_list']

        if pd.isna(aspects_str) or aspects_str.strip() == '' or \
           pd.isna(sentiment_str) or sentiment_str.strip() == '':
            continue
        #arabic comma dont forget
        individual_aspects = [a.strip() for a in aspects_str.split('،')]
        # print(individual_aspects)

        try:
            individual_sentiments = strip_sentiment_list(sentiment_str)
        except ValueError:
            print(f"Warning: Could not parse sentiment string '{sentiment_str}' in row {index}. Skipping.")
            continue

        # Ensure that the number of aspects matches the number of sentiments
        if len(individual_aspects) == len(individual_sentiments):
            for aspect, sentiment in zip(individual_aspects, individual_sentiments):
                aspect_sentiment_pairs.append({'Aspect': aspect, 'Sentiment': sentiment})
        else:
            # Handle cases where lengths might not match, or simply skip
            print(f"Warning: Mismatch in aspect/sentiment count in row {index}. Aspects: {len(individual_aspects)}, Sentiments: {len(individual_sentiments)}. Skipping.")

    return pd.DataFrame(aspect_sentiment_pairs)


def plot_heatmap(df:pd.DataFrame):

  # Extract the aspect-sentiment pairs
  aspect_sentiment_df = extract_aspect_sentiment_pairs(df)

  # Count occurrences of each aspect for each sentiment
  heatmap_data = aspect_sentiment_df.groupby(['Aspect', 'Sentiment']).size().unstack(fill_value=0)

  # If you want to focus on top N aspects, uncomment and adjust the following lines:
  top_n_aspects = 10 # Number of top aspects to display
  top_aspects_list = heatmap_data.sum(axis=1).nlargest(top_n_aspects).index
  heatmap_data = heatmap_data.loc[top_aspects_list]

  # Ensure all sentiment columns are present, even if some are all zeros
  for col in [-1, 0, 1]:
      if col not in heatmap_data.columns:
          heatmap_data[col] = 0
  heatmap_data = heatmap_data[[-1, 0, 1]] # Order the columns correctly



  reshaped_aspect_labels = [get_display(arabic_reshaper.reshape(label)) for label in heatmap_data.index]

  plt.figure(figsize=(12, max(6, len(heatmap_data) * 0.4))) # Adjust figure size dynamically
  sns.heatmap(heatmap_data,
              annot=True, # Annotate cells with the numeric value
              fmt='d',    # Format annotations as integers
              cmap='viridis', # Choose a colormap
              linewidths=.5, # Add lines between cells
              linecolor='black',
              yticklabels=reshaped_aspect_labels)

  plt.title('Heatmap of Aspect Sentiments (Counts)')
  plt.xlabel('Sentiment Score')
  plt.ylabel('Aspect')

  # Customize x-axis labels for clarity
  plt.xticks(ticks=[0.5, 1.5, 2.5], labels=['Negative (-1)', 'Neutral (0)', 'Positive (1)'])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display
from collections import Counter

# Function to extract individual aspects, reshape them, and count their frequencies
def get_reshaped_aspect_word_frequencies(df: pd.DataFrame):
    all_raw_aspect_words = []
    for aspects_str in df['Aspects_clean'].dropna():
        # Split by both Arabic and English commas, then strip whitespace
        individual_aspects = [a.strip() for a in aspects_str.replace(',', '،').split('،')]
        # Filter out any empty strings
        all_raw_aspect_words.extend([aspect for aspect in individual_aspects if aspect])

    reshaper = ArabicReshaper(configuration={'language': 'Arabic'})
    reshaped_logical_words = [] # Renamed for clarity
    for word in all_raw_aspect_words:
        # Reshape the word. WordCloud with a proper font will handle BiDi.
        reshaped_word = reshaper.reshape(word)
        reshaped_logical_words.append(reshaped_word)

    # Print a few examples to verify
    print("\n--- Verifying individual Arabic word rendering ---")
    print(f"Original (first 5): {[word for word in all_raw_aspect_words[:5] if word.strip()]}")
    print(f"Reshaped (logical order) (first 5): {[word for word in reshaped_logical_words[:5] if word.strip()]}")
    print("--------------------------------------------------\n")

    # Count frequencies of these reshaped and display-ready words
    word_frequencies = Counter(reshaped_logical_words)
    return word_frequencies
def plot_word_cloud(df: pd.DataFrame,domain:str):
  # Get the frequencies of reshaped aspect words
  word_freqs = get_reshaped_aspect_word_frequencies(df)

  # Define a font path that supports Arabic characters (trying Amiri this time)
  font_path_arabic = '/usr/share/fonts/truetype/amiri/Amiri-Regular.ttf'

  # Generate the word cloud from frequencies
  wordcloud = WordCloud(
      width=800,
      height=400,
      background_color='white',
      font_path=font_path_arabic, # Specify the Arabic font
      mode='RGB', # Use RGB mode
      max_words=100
  ).generate_from_frequencies(word_freqs)

  # Display the word cloud
  plt.figure(figsize=(10, 5))
  plt.imshow(wordcloud, interpolation='bilinear')
  plt.axis('off')
  plt.title(domain)
  plt.show()
```

```python
import networkx as nx
import plotly.graph_objects as go
from itertools import combinations
from collections import Counter
import ast

# Get unique domain names from the original combined_data.csv
df_full = pd.read_csv("combined_data.csv", encoding="utf-8")
unique_domains_for_graphs = df_full['domain_name'].unique()

for domain in unique_domains_for_graphs:
    print(f"\nGenerating Aspect Co-occurrence Network for domain: {domain}")

    # Load a fresh copy of the data for each domain processing
    domain_df = pd.read_csv("combined_data.csv", encoding="utf-8")
    domain_df = domain_df[domain_df['domain_name'] == domain].copy()

    # Safely convert the string representation of lists in 'Aspects_list' to actual lists
    domain_df['Aspects_list'] = domain_df['Aspects_list'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else None)

    # Drop rows where Aspects_list could not be parsed or are None
    domain_df = domain_df.dropna(subset=['Aspects_list'])

    # Explode the list of aspects into separate rows
    domain_df = domain_df.explode('Aspects_list')

    # Strip any remaining whitespace from the exploded aspects
    domain_df['Aspects_list'] = domain_df['Aspects_list'].str.strip()

    # Remove any empty strings from Aspects_list
    domain_df = domain_df[domain_df['Aspects_list'] != '']

    # Group back per original sentence to get aspects that co-occur in the same sentence
    grouped_sentences_domain = domain_df.groupby(domain_df.index)['Aspects_list'].apply(list)

    pairs_domain = []
    for aspect_list in grouped_sentences_domain:
        pairs_domain.extend(list(combinations(set(aspect_list), 2)))

    pair_counts_domain = Counter(pairs_domain)

    # Implement dynamic threshold for strong connections
    if not pair_counts_domain:
        dynamic_threshold = 1 # If no pairs, effectively no threshold needed
    else:
        max_count = max(pair_counts_domain.values())
        # Set threshold to at least 2, or 10% of max count, whichever is higher
        dynamic_threshold = max(2, int(max_count * 0.1))

    print(f"Dynamic threshold for {domain}: {dynamic_threshold}")

    strong_pair_counts_domain = {k: v for k, v in pair_counts_domain.items() if v >= dynamic_threshold}

    G_domain = nx.Graph()

    for (a1, a2), weight in strong_pair_counts_domain.items():
        G_domain.add_edge(a1, a2, weight=weight)

    # Only proceed if the graph is not empty
    if G_domain.number_of_nodes() > 0:
        pos_domain = nx.spring_layout(G_domain, k=0.5)

        edge_x = []
        edge_y = []

        for edge in G_domain.edges():
            x0, y0 = pos_domain[edge[0]]
            x1, y1 = pos_domain[edge[1]]
            edge_x += [x0, x1, None]
            edge_y += [y0, y1, None]

        edge_trace = go.Scatter(
            x=edge_x, y=edge_y,
            line=dict(width=0.5),
            hoverinfo='none',
            mode='lines'
        )

        node_x = []
        node_y = []
        text = []

        for node in G_domain.nodes():
            x, y = pos_domain[node]
            node_x.append(x)
            node_y.append(y)
            text.append(node)

        node_trace = go.Scatter(
            x=node_x, y=node_y,
            mode='markers+text',
            text=text,
            textposition="top center",
            hoverinfo='text',
            marker=dict(size=10)
        )

        fig_domain = go.Figure(data=[edge_trace, node_trace],
                        layout=go.Layout(title=f"Aspect Co-occurrence Network for {domain}"))

        fig_domain.show()
    else:
        print(f"No strong aspect co-occurrences found for domain: {domain} to form a graph with threshold {dynamic_threshold}.")
```
